```
import gym
import numpy as np
env = gym.make('CartPole-v0') #建立環境

def play():
    
    #建立策略梯度
    learning_rate = -0.00001
    weight = np.random.rand(4)
    action = env.action_space.sample() #第一次的動作
    observation = env.reset()
    #建立簡單策略網路
    for iteration in range(1000): 
        observation, reward, done, info = env.step(action)
        weight_sum = np.dot(weight, observation) #做簡單的內積
        policy = 1 / (1 + np.exp(-weight_sum))  #透過Ｓ函數計算動作
        if policy > 0.5: #擇中取0.5
            action = 1
        else:
            action = 0
            #策略梯度公式=> Theta=步長*Theta*獎勵
            weight += learning_rate*(1-policy)*np.transpose(-observation)*reward
    print(weight)
    for i in range(200): #Episode次數
        observation = env.reset()
        rewards = 0
        for t in range(2500): #每段時間做的回饋
            weight_sum = np.dot(weight, observation)
            policy = 1 / (1 + np.exp(-weight_sum))   
            if policy > 0.5:
                action = 1
            else:
                action = 0             
            observation, reward, done, info = env.step(action) #回饋獎勵、狀態
            rewards += reward
            if done:
                print('Episode結束 {} 次, 總獎勵 {}'.format(t+1, rewards))
                break
            #env.render()
    env.close()
    
play()
```
