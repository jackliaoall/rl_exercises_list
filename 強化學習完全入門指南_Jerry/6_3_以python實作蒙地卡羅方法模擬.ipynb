```
from numpy.random import normal
from matplotlib import pyplot as plt

# 產生蒙地卡羅樣本數
s = [10, 20, 100, 500, 1000, 3000]

for i in range(len(s)):
	# 產生樣本
	sample_mc = normal(100, 10, s[i]) #產生平均值100，標準差10的樣本
	# 繪製圖表
	plt.subplot(3, 3, i+1)
	plt.hist(sample_mc, bins=30)
	plt.title('%d data' % s[i])
	plt.xticks([])
plt.show()
```
