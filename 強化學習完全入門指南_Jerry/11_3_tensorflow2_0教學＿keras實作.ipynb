```
# Commented out IPython magic to ensure Python compatibility.
# %tensorflow_version 2.x

import tensorflow as tf
from tensorflow import keras

#建立基本神經網路架構
from tensorflow.keras import layers
model = tf.keras.Sequential()
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(128, activation='relu'))
model.add(layers.Dense(3, activation='softmax'))

#設定隱藏層的參數
model.add(layers.Dense(64, activation='sigmoid'))
#或是以下寫法
model.add(layers.Dense(64, activation=tf.keras.activations.sigmoid))
#加上正規化L1 
model.add(layers.Dense(64, kernel_regularizer=tf.keras.regularizers.l1(0.01)))
model.add(layers.Dense(64, bias_regularizer=tf.keras.regularizers.l2(0.01)))

#訓練模型與驗證
model = tf.keras.Sequential([
layers.Dense(64, activation='relu', input_shape=(32,)),
layers.Dense(64, activation='relu'),
layers.Dense(10, activation='softmax')])

model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])


#數值型題目衡量標準
model.compile(optimizer=tf.keras.optimizers.Adam(0.01),
              loss='mse',       #衡量標準MSE
              metrics=['mae'])  #儲存的參數

#類別型題目衡量標準
model.compile(optimizer=tf.keras.optimizers.RMSprop(0.01),
              loss=tf.keras.losses.CategoricalCrossentropy(),
              metrics=[tf.keras.metrics.CategoricalAccuracy()])

#儲存權重參數
model = tf.keras.Sequential([
layers.Dense(64, activation='relu', input_shape=(32,)),
layers.Dense(10, activation='softmax')])

model.compile(optimizer=tf.keras.optimizers.Adam(0.001),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
#儲存權重的檢查點
model.save_weights('./weights/my_model')

!ls

#還原權重檢查點（需要模型是同一種架構才行）
model.load_weights('./weights/my_model')

#儲存成HDF5格式
model.save_weights('my_model.h5', save_format='h5')

#還原HDF5格式
model.load_weights('my_model.h5')
```
