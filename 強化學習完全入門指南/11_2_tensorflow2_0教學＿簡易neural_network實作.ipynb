```
import numpy as np

def NN(m1, m2, w1, w2, b):
    z = m1 * w1 + m2 * w2 + b
    return sigmoid(z)

def sigmoid(x):
    return 1/(1 + np.exp(-x))

w1 = np.random.randn() 
w2 = np.random.randn() 
b = np.random.randn()

NN(3,1.5,w1,w2,b)

phrases = ['看起來像', '我猜是', '我想是', '可能是', '看起來像是', '猜測....']
data = [[3,1.5,1],[2,1,0],[4,1.5,1],[3,1,0],[3.5,5,1],[2,0.5,0],[5.5,1,1],[1,1,0]]
rand_data = data[np.random.randint(len(data))] #隨機產生len(data)-1的index
m1 = rand_data[0]
m2 = rand_data[1]

prediction = NN(m1,m2,w1,w2,b)
prediction_text = ['雞腿飯','義大利麵'][int(np.round(prediction))] #取四捨五入後的數值
phrase = np.random.choice(phrases) + "" + prediction_text #隨機取phrases
o =(phrase + "\n真的是"+ ['雞腿飯','義大利麵'][rand_data[2]])
print(o)
```
